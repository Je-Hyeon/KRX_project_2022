# [기업코드 : 시장] 파일을 만들어보자

```python
import numpy as np
import pandas as pd
import os

list_dir = os.listdir("Data/KIND데이터/코스피+코스닥/")
df_dict = {}

for name in list_dir[:-2]:
    df_dict["df"+str(name[:-14])] = pd.read_csv("Data/KIND데이터/코스피+코스닥/{}".format(name), index_col=0)
```

```python
concat_df = pd.DataFrame()

for key,item in df_dict.items():
    concat_df = pd.concat([concat_df,item[["기업공시코드","시장"]]])

df = concat_df.drop_duplicates().reset_index(drop=True)
```

```python
df.to_csv("기업코드_시장.csv")
```
